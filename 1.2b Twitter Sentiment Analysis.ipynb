```python
import numpy as np
import pandas as pd
import time
import matplotlib.pyplot as plt
import squarify
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from datetime import date
import glob
import os
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
vader = SentimentIntensityAnalyzer()
```

```python
vader.lexicon.update(new_words)
```

```python
# Change working directory for Twitter posts
os.chdir('../data/datafiles/twitter/')
```

```python
# Example of use case
'''
Input: String
Output: Dict {}
'''
sentence = 'The acting was good , but the movie could have been better'
vader.polarity_scores(sentence)
```

```python
# Perform sentiment analysis on Twitter post text
# Extract both the polarity scores and the compound sentiment score
# Save the CSV files with the calculated sentiment scores
for file in os.listdir()[1:]:
    print(f'Starting {file} analysis...')
    tmp = pd.read_csv(file, index_col = 0)
    tmp['polarity_scores'] = tmp['tweet'].apply(lambda text: vader.polarity_scores(str(text)))
    tmp['compound_score']  = tmp['polarity_scores'].apply(lambda score_dict: score_dict['compound'])
    # tmp.to_csv(file)
    print(f'Completed {file} analysis.')
```
